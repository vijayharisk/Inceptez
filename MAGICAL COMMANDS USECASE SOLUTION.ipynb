```python
from pyspark.sql.functions import col

# Read CSV into DataFrame
df1 = spark.read.csv(
    "dbfs:/Volumes/workspace/default/usage_metrics/mobile_os_usage.csv",
    header=True,
    inferSchema=True
)

df1_clean = df1.select(
    col("Date"),
    col("Mobile Operating System").alias("Mobile_Operating_System"),
    col("Percent of Usage").alias("Percent_of_Usage")
)
# Save as a managed table
df1_clean.write.saveAsTable("dataframe_table")
```
